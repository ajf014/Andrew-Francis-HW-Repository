```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Run a length of the unique variables of the Item ID column to get total number of unique items
#Name the variable unique items
unique_items = len(purchase_data["Item ID"].unique())
unique_items
#Get the average price by looking at the Price column and running the .mean function.
average_price = purchase_data["Price"].mean()
average_price
#Use .count() to get the number of purchases made from the Price column.
purchases = purchase_data["Price"].count()
purchases
#Use the .sum to get the total value from the the Price column
total_revenue = purchase_data["Price"].sum()
total_revenue

#Create a datafranme where I create new names for the columns and use the variables a created above.
Summary = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [average_price],
                        "Number of Purchases": [purchases], "Total Revenue": [total_revenue]})

#Formatted Section
Summary["Average Price"] = Summary["Average Price"].map("${:,.2f}".format)
Summary["Total Revenue"] = Summary["Total Revenue"].map("${:,.2f}".format)

Summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed


```python
# use a .loc to find the information for the Gender, SN, and Age columns.
player_demographics = purchase_data.loc[:, ["Gender", "SN","Age"]]
player_demographics

# Drop the duplicates
player_demographics = player_demographics.drop_duplicates()

#Count the number of people by each gender
gender = player_demographics["Gender"].value_counts()
gender

#Find the percentage of players by dividing gender variable created above by screen name multiplied by 100.
# The gender variable splits up the calculation by gender for you
percentage = (gender / SN * 100)

#Create dataframe to display all data.
gender_summary = pd.DataFrame({"Total Count": gender,"Percentage of Players": percentage})
gender_summary

#Formatted Section
gender_summary_formatted = gender_summary
gender_summary_formatted["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:,.2f}".format)
gender_summary_formatted
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Take the purchase_data, group by SN because that has all the unique values and use the describe command to get the columns you need
summary = purchase_data.groupby("SN")["Price"].describe()
summary

# Create data frame with the columns you created from the summary directly above
top_spenders = pd.DataFrame({"Purchase Count": summary["count"],
                            "Average Purchase Price": summary["mean"],
                            "Total Purchase Value": summary["mean"]*summary["count"]
                        })
#Sort the top spenders
top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)
top_spenders.head()

#Formatted
top_spenders_formatted = top_spenders
top_spenders_formatted["Purchase Count"] = top_spenders["Purchase Count"].map("{:,.0f}".format)
top_spenders_formatted["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_formatted["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:,.2f}".format)

top_spenders_formatted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create new variable with the following columns:
Popular_new = purchase_data.drop(columns=['SN', 'Age', 'Gender', 'Purchase ID', 'Age Group'])

# Group by Item ID and Item name, then describe by Price column
Popular_new = Popular_new.groupby(['Item ID', 'Item Name'])['Price'].describe()
# Drop the columns I don't need
Popular_new = Popular_new.drop(columns=['std', 'min', '25%', '50%', '75%','max'])
# Get total purchase value by multiplyin count and mean
Popular_new["Total Purchase Value"] = Popular_new["count"] * Popular_new["mean"]
# Rename the columns
Popular_new = Popular_new.rename(columns={"count": "Purchase Count", "mean": "Item Price"})

# Formatted
Popular_formatted = Popular_new.sort_values("Purchase Count", ascending=False)
Popular_formatted["Item Price"] = Popular_formatted["Item Price"].map("${:.2f}".format)
Popular_formatted["Purchase Count"] = Popular_formatted["Purchase Count"].map("{:.0f}".format)

Popular_formatted.head()
```
